## Блок 02 - Агрегированные группировки датафреймов

```python
# Импорт библиотеки Pandas, метода parse из библиотеки urllib и открытие файла module_04_ad_campaigns.xlsx
import pandas as pd
from urllib import parse

ad_stats = pd.read_excel('module_04_ad_campaigns.xlsx')

# Переименование столбцов из кириллицы в латиницу
ad_stats.rename(columns={'Название группы': 'group', 'Фраза (с минус-словами)': 'phrase', 
                         'Продуктивность': 'effect', 'ID объявления': 'ad_id', 
                         'Заголовок': 'title','Текст': 'text', 'Ссылка': 'link'}, inplace=True)

# Функция для работы с датафреймом
def campaign_name(row):
    parsed = parse.urlsplit(row['link'])
    params_dict = parse.parse_qs(parsed.query)
    return params_dict['utm_campaign'][0]

ad_stats['campaign'] = ad_stats.apply(campaign_name, axis = 1)
ad_stats.head()
```

```python
def effect_class(row):
    if row['effect'] <= 3:
        return 'Низкий'
    if row['effect'] <= 7:
        return 'Средний'
    return 'Высокий'

ad_stats['effect_class'] = ad_stats.apply(effect_class, axis = 1)
ad_stats.head()
```

```python
# Создаём сводную таблицу в строках "Кампании" в столбцах "Эффект от рекламы"
# содержимое ячеек функция(ex. sum or etc...) применёная к значению столбца 'effect' в исходной таблице
ad_stats.pivot_table(index='campaign', columns='effect_class', values='effect', aggfunc='count', fill_value='0')
```

## Блок 04 - Агрегированные столбцы датафреймов

```python
# Пишем алгоритм, показывающий распределение оценок для заданного фильма movieId

# Импорт библиотеки Pandas, открытие файла module_05_ratings.csv

import pandas as pd

ratings = pd.read_csv('module_05_ratings.csv')
ratings.head()
```

```python
ratings_pivot = ratings.pivot_table(index=['movieId','rating'], values='timestamp', aggfunc='count').reset_index()
ratings_pivot.head(10)
```

```python
ratings_pivot.groupby('movieId').timestamp.transform(lambda x: x)
```

```python
# Добавляем в лямбда функуию sum(x) и сохраняем результат в столбце sum 
ratings_pivot['sum'] = ratings_pivot.groupby('movieId').timestamp.transform(lambda x: sum(x))
ratings_pivot.head(10)
```

```python
# Получаем долю оценок каждого рейтинга для каждого фильма
ratings_pivot['share'] = ratings_pivot['timestamp'] / ratings_pivot['sum']
ratings_pivot.head(10)
```

```python
# Вычислить сумму столбца share для movieId = 1
ratings_pivot[ratings_pivot['movieId'] == 1]['share'].sum()
```

```python
# написать алгоритм, который классифицирует эти оценки по более практичной шкале:
# · оценка 2 и меньше — низкий рейтинг;
# · оценка 4 и меньше — средний рейтинг;
# · оценка 4.5 и 5 — высокий рейтинг.

import pandas as pd

userfriendly_ratings = pd.read_csv('module_05_ratings.csv')

def rating_class(row):
    if row['rating'] <= 2:
        return 'Низкий'
    if row['rating'] <= 4:
        return 'Средний'
    return 'Высокий'

userfriendly_ratings['rating_class'] = userfriendly_ratings.apply(rating_class, axis = 1)
userfriendly_ratings.head(10)
```

```python
userfriendly_ratings_pivot = userfriendly_ratings.pivot_table(
    index=['movieId', 'rating_class'], values='timestamp', aggfunc='count').reset_index()
userfriendly_ratings_pivot.head(10)
```

```python
userfriendly_ratings_pivot['sum'] = userfriendly_ratings_pivot.groupby('movieId').timestamp.transform(lambda x: sum(x))
userfriendly_ratings_pivot.head(10)
```

```python
userfriendly_ratings_pivot['share'] = round(userfriendly_ratings_pivot['timestamp'] / userfriendly_ratings_pivot['sum'], 2)
userfriendly_ratings_pivot.head(10)
```
