### Блок 03 - Склеивание датафреймов через concatenate

```python
df1 = pd.DataFrame({'A': ['A0', 'A1', 'A2', 'A3'],
'B': ['B0', 'B1', 'B2', 'B3'],
'C': ['C0', 'C1', 'C2', 'C3'],
'D': ['D0', 'D1', 'D2', 'D3']})

df2 = pd.DataFrame({'A': ['A4', 'A5', 'A6', 'A7'],
'B': ['B4', 'B5', 'B6', 'B7'],
'C': ['C4', 'C5', 'C6', 'C7'],
'E': ['D4', 'D5', 'D6', 'D7']})

# У итогового датафрейма значения индекса дублируются. Каким методом его можно обновить?
pd.concat([df1, df2]).reset_index()
```

```python
# Объединение файлов в dataframe

# Создаём dataframe
data = pd.DataFrame(columns=['userId', 'movieId', 'rating', 'timestamp'])
```

```python
# Получаем список файлов для чтения и фильтруем его
files = os.listdir('data')
new_files = []
for file in files:
    if (('ratings_' in file) & ('txt' in file)):
        new_files.append(file)
files = new_files
files
```

```python
# Используем временный dataframe temp для чтения содеержимого файлов
for filename in files:
    temp = pd.read_csv(os.path.join('data', filename), names=['userId', 'movieId', 'rating', 'timestamp'])
    # Используем метод concat для добавления содержимого очередного файла в основной dataframe
    data = pd.concat([data, temp])
```
